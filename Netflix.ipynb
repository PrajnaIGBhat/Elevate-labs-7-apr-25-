```python
import pandas as pd
```

```python
data=pd.read_csv("Netflix data.csv")
data
```

#Identify and Handle Missing Values

```python
data.isnull().sum()
```

```python
#Filling the nullvalues
```

```python
columns_to_fill = ['director', 'cast', 'country','date_added','duration']

data[columns_to_fill] = data[columns_to_fill].fillna('Unknown')
```

```python
data['rating']=data['rating'].fillna(value='not rated')
data
```

```python
data.isnull().sum()
```

#Remove Duplicate Rows

```python
data.drop_duplicates()
```

```python
data.duplicated().sum()
```

```python
#Standardize Text Values
```

```python
for col in data.select_dtypes(include='object').columns:
    data[col] = data[col].str.strip().str.title()
```

```python
data
```

# Convert Date Format (date_added) to dd-mm-yyyy

```python
data['date_added'] = pd.to_datetime(data['date_added'], errors='coerce').dt.strftime('%d-%m-%Y')
```

```python
data
```

#Clean Column Headers

```python
data.columns = data.columns.str.lower().str.strip().str.replace(' ', '_')
data
```
